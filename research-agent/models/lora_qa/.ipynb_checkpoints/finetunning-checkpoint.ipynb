```python
from datasets import load_dataset

train_dataset = load_dataset("json", data_files="train_dataset.jsonl")["train"]
print(train_dataset[1])
```

```python
from transformers import Trainer, TrainingArguments
```

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, TrainingArguments, Trainer, DataCollatorForLanguageModeling
from peft import LoraConfig, get_peft_model
# ----------------------
# Step 2: Base model + tokenizer
# ----------------------
model_id = "meta-llama/Meta-Llama-3-8B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_id,
    load_in_4bit=True,  # saves VRAM
    device_map="auto"
)

# ----------------------
# Step 3: Apply LoRA
# ----------------------
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    target_modules=["q_proj", "v_proj"],  # attention projections in LLaMA
    lora_dropout=0.05,
    bias="none",
    task_type="CAUSAL_LM",
)
model = get_peft_model(model, lora_config)

# ----------------------
# Step 4: Tokenization
# ----------------------
def tokenize(batch):
    prompt = batch["instruction"] + batch.get("input", "")
    answer = batch["output"]
    text = f"Question: {prompt}\nAnswer: {answer}"
    return tokenizer(text, truncation=True, padding="max_length", max_length=512)

train_dataset = train_dataset.map(tokenize, batched=True)

data_collator = DataCollatorForLanguageModeling(tokenizer, mlm=False)
```
